```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sales distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=50, kde=True)
plt.title("Sales Distribution")
plt.show()

# Profit distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Profit'], bins=50, kde=True)
plt.title("Profit Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Category', y='Sales', data=df, estimator=sum)
plt.title("Total Sales by Category")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Region', y='Profit', data=df, estimator=sum)
plt.title("Total Profit by Region")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sub_cat = df.groupby('Sub-Category').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sns.barplot(x='Sub-Category', y='Sales', data=sub_cat)
plt.title("Total Sales by Sub-Category")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(x='Sub-Category', y='Profit', data=sub_cat)
plt.title("Total Profit by Sub-Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Discount', y='Profit', hue='Category', data=df)
plt.title("Discount vs Profit by Category")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Segment', y='Profit', data=df, estimator=sum)
plt.title("Total Profit by Segment")
plt.show()
```

```python
pivot = df.pivot_table(values='Sales', index='Region', columns='Category', aggfunc='sum')
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Sales Heatmap: Region vs Category")
plt.show()
```

```python
import pandas as pd

# Date columns ko datetime me convert karna
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Ab naye columns create kar sakte ho
df['Order_Year'] = df['Order Date'].dt.year
df['Order_Month'] = df['Order Date'].dt.month
df['Order_Month_Year'] = df['Order Date'].dt.to_period('M')

# Shipping days calculate
df['Shipping_Days'] = (df['Ship Date'] - df['Order Date']).dt.days

# Profit Margin
df['Profit_Margin'] = df['Profit'] / df['Sales']

# Check first 5 rows
df.head()

```

```python
yearly_sales = df.groupby('Order_Year')['Sales'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.lineplot(x='Order_Year', y='Sales', data=yearly_sales, marker='o')
plt.title("Year-wise Total Sales")
plt.show()
```

```python
# Month-Year ko string me convert
df['Order_Month_Year_str'] = df['Order_Month_Year'].astype(str)

# Month-Year-wise Sales Trend
monthly_sales = df.groupby('Order_Month_Year_str')['Sales'].sum().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(x='Order_Month_Year_str', y='Sales', data=monthly_sales, marker='o')
plt.xticks(rotation=45)
plt.title("Month-Year-wise Sales Trend")
plt.show()
```

```python
category_margin = df.groupby('Category')['Profit_Margin'].mean().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(x='Category', y='Profit_Margin', data=category_margin)
plt.title("Average Profit Margin by Category")
plt.show()
```
